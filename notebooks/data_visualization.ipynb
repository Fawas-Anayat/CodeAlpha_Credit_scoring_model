## Data Visualization
This file is specific for the visualization of the data that I am going to train my model on

### importing imporatant libraries and modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the datset

```python
df=pd.read_csv('../data/cleaned_data.csv')
df.head()
```

```python
df.drop(df.columns[0:1],axis=1)
```

### Visualizing the data

```python
plt.figure(figsize=(12,6))
plt.scatter(df['monthly_income'],df['monthly_expense_total'],alpha=0.6,color='blue',edgecolors='black',s=30)
plt.title('Monthly Income VS Monthly Expense')
plt.xlabel('Monthly Income')
plt.ylabel('Monthly expense')
plt.grid()
plt.savefig('../plots/income_vs_expenses.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,5))
value_counts=(df['income_type']).value_counts()
plt.pie(value_counts,labels=value_counts.index,startangle=90,autopct='%1.1f%%')
plt.title('Distribution w.r.t income type')
plt.savefig('../plots/distribution_of_income_types.png',dpi=350,bbox_inches='tight')
plt.show()
```

```python
selected_col=['savings_rate','debt_to_income_ratio','transaction_count','financial_advice_score']
df_selected=df[selected_col]

corr = df_selected.corr(numeric_only=True)  # ensures only numbers are included

# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Correlation Heatmap")
plt.savefig('../plots/correlation_heatmap.png')
plt.show()
```
